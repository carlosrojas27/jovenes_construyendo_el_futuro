```python
# Selección 'if', 'elif' y 'else'
if 1>5:
  print('verdadero')
else:
  print('Falso')
# siempre se hace una comparación, una pregunta.
# Si diez es mayor a cinco entonces imprimes verdadero, de lo contrario si diez no es mayor a cinco imprimes falso
```

```python
print('Esta es una calculadora')
print('Elije el tipo de operación del menú')
print('1. Sumar')
print('2. Restar')
print('3. Multiplicar')
print('4. Dividir')
res=input('¿qué quieres hacer? ')
num1=float(input('Escribe el primer numero: '))
num2=float(input('Escribe el segundo numero: '))
if res=='sumar':
  print('{} + {} = {}'.format(num1,num2,num1+num2)) 
elif res=='restar':
  print('{} - {} = {}'.format(num1,num2,num1-num2)) 
elif res=='multiplicar':
  print('{} * {} = {}'.format(num1,num2,num1*num2)) 
elif res=='dividir':
  print('{} / {} = {}'.format(num1,num2,num1/num2)) 
else:
  print('comando no válido')

# pass, palabra reservada que indica no hacer nada.
```
